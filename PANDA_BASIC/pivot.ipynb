```python
ex_pivot=pd.pivot_table(ex_data,values='Value',index=['Customer', 'Quantity'],columns='Product')
ex_pivot
```

```python
ex_pivot=pd.pivot_table(ex_data,values='Value',index=['Customer'],columns=['Product', 'Quantity'],fill_value=0)
# bydefult  aggfunc=mean
ex_pivot
```
