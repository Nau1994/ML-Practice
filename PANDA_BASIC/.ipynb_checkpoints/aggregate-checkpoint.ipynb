```python
import pandas as pd
from pandas import DataFrame
order=pd.read_excel("SuperstoreSales.xls",0)
order.columns
```

```python
order.groupby('Customer Segment').Sales.sum()
```

```python
order.groupby('Customer Segment').Sales.mean()
```

```python
order.groupby('Customer Segment').Sales.max()
```

```python
order.groupby('Customer Segment').Sales.agg(["sum","mean","max"])
```
